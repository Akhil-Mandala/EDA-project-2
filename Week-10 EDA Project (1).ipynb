# Find out the distribution of sales over different times of the day (Morning, Afternoon, Evening, Night) and answer the following for each city?

```python
#defining part of the day weather it's morning,afternoon,evening and night by giving the condition forming hours into 4 parts. 
def get_part_of_day(hour):
    return(
        'morning'
    if 6<=hour<12
    else'afternoon'
    if 12<=hour<=17
    else 'evening'
    if 17<=hour<=20
    else'night'
    )
df['Part_of_day']= df['Hour'].apply(get_part_of_day)
```

```python
df.head(5)
```

## i)What products are mostly sold at different times of the day?

```python
#To know which products are sold mostly in different times plotting a count plot using seaborn library giving part_of_the_day on X-axis
# products ooh hue and pallette as deep
plt.figure(figsize=(25,40))
sns.countplot(hue=df['Product'],x=df['Part_of_day'],palette='deep')
plt.show()
```

### Afternoon has the highest on day Lighting Charging Cable and USB-C charging cable are mostly sold.

## ii)Which time of day has the most sales?

```python
#To know which time of day has sold mostly in different times plotting a count plot using seaborn library giving city on X-axis and
# part_of_day on hue and pallette as hot.
plt.figure(figsize=(30,40))
sns.countplot(hue=df['Part_of_day'],x=df['City'],palette='hot')
plt.show()
```

# Find out the pair of products that are most sold together. For example, generally, we would expect a person who buys an iPhone 11 to also buy the Apple Airpods together with it. Identify such pairs of products and suggest what promotional offers should the retail company come up with based on your outcome.

```python
df.head()
```

```python
df1=df.drop_duplicates(keep='first')
df1
```

```python
#Now we want to create a column called “Product_together” that contain example iPhone and Wired Headphone at the same line. We’re gonna use the .transform() method to join values from two rows into a single row.
df1['Product_together']=df1.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df1
```

```python
#It’s good, but we have an issue here. We have the same order at least twice because we did merge them in every situation in groupby without dropping the duplicate values. Now let’s drop the rows with duplicate values.
df2=df1[['Order ID','Product_together']].drop_duplicates()
df2.head()
```

```python
#Importing libraries
#Now, we need to count the pair of products. We need new libraries because they have all we need to count all the combination of products bundle. We’re gonna use itertools and collections libraries.
from itertools import combinations
from collections import Counter

count = Counter()

for row in df1['Product_together']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)    
```

### We can see the most often sold products  together are'iPhone', 'Lightning Charging Cable'.
